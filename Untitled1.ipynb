```python
import cv2
import numpy as np
```

```python
mask = np.load('with_mask.npy')
without_mask = np.load('without_mask.npy')
```

```python
mask.shape, without_mask.shape
```

```python
mask[0]
```

```python
mask = mask.reshape(200, 50*50*3)
without_mask = without_mask.reshape(200,50*50*3)
```

```python
mask.shape
```

```python
X = np.r_[mask,without_mask]
X.shape
```

```python
labels = np.zeros(X.shape[0])
```

```python
labels[200:] = 1.0
```

```python
#svc = support vector classifier
#svm = support vector machine
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.25)
```

```python
X_train.shape
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
x_train = pca.fit_transform(X_train)
x_train[0]
```

```python
x_train.shape
```

```python
svc = SVC()
svc.fit(x_train,y_train)
```

```python
x_test = pca.fit_transform(X_test)
y_pred = svc.predict(x_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
import matplotlib.pyplot as plt
from sklearn import metrics
fpr, tpr, _ = metrics.roc_curve(y_test, y_pred)
plt.plot(fpr,tpr)
plt.title('ROC Curve')
plt.ylabel('True Positive rate')
plt.xlabel('False Positive rate')
plt.show()
```

```python
precision, recall, thresold = metrics.precision_recall_curve(y_test,y_pred)
plt.plot(precision,recall)
plt.title('PR -Curve')
plt.ylabel('Precision')
plt.xlabel('Recall')
plt.show()
```

```python
haar_data = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
name = {0:'mask', 1:'NO mask'}
```

```python
cap = cv2.VideoCapture(0)
font = cv2.FONT_HERSHEY_COMPLEX
while True:
    flag, img = cap.read()
    if flag:
        faces = haar_data.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img,(x,y),(x+w,y+h),(27,0,255), 4)
            face = img[y:y+h, x:x+w, :]
            face = cv2.resize(face,(50,50))
            face = face.reshape(1,-1)
            face = pca.transform(face)
            pred = svc.predict(face)
            n = name[int(pred)]
            cv2.putText(img,n,(x,y),font, 1,(224,250,250),2)
        cv2.imshow('result',img)
        if cv2.waitKey(2)==27:
            break

            
cap.release()
cv2.destroyAllWindows()
```
